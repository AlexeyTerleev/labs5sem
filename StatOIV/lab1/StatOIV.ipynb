```python
from zipfile import ZipFile
import pandas as pd
```

```python
with ZipFile('spaceship-titanic.zip', 'r') as zObject:
    zObject.extractall(path="dataset")

df = pd.read_csv("dataset/train.csv")
```

```python
df.head(5)
```

```python
coloums_target = ["Transported", "HomePlanet", "CryoSleep", "Age", "VIP"]
filtered_df = df[coloums_target].copy()
```

```python
filtered_df.head(5)
```

```python
home_planet_dict = {'Europa': 0, 'Earth': 1, 'Mars': 3}
filtered_df["HomePlanet"] = filtered_df["HomePlanet"].apply(lambda x: home_planet_dict.get(x, None))
filtered_df["Age"] = filtered_df["Age"].fillna(filtered_df["Age"].median())
```

```python
for coloumn in ["CryoSleep", "VIP"]:
  filtered_df[coloumn] = filtered_df[coloumn].astype(bool)
```

```python
filtered_df.head(5)
```

```python
filtered_df.to_csv("filtered_df.csv")
```
